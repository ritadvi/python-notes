## Seaborn  
Categorical plots

```python
import seaborn as sns
import matplotlib as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

**Bar plot with Estimator**

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

**Bar plot with count**  
y value is by default quantity

```python
sns.countplot(x='sex',data=tips)
```

**Boxplot**  
Compare distribution of categorical data

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

**Violin plot**  

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

**Strip plot**

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
# Use jitter to add random noise and separate the stacked points
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

**Swarm plot**  
Violin plot + strip plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

**Factor plot**  
More general, specify kind

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```
