# Seaborn
## Matrix plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

**Correlations**

```python
tc = tips.corr()
tc
```

**Heat Map** 

```python
# Add _r at the end of cmap to reverse color scale, in this case coolwarm_r
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

**Pivot Table**

```python
flights
```

```python
# Pivot table - total passengers per month/year
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

**Heat Map 2**  
LineColor, LineWidth

```python
sns.heatmap(fp, cmap='magma_r', linecolor='white', linewidth=3 )
```

**Cluster Map**  
Shows relationships between most similar months/years.  
Dimensions are re ordered based on their similarities.

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```
