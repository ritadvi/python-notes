# Matplotlib  
## Appearance

```python
import matplotlib.pyplot as plt
import numpy as np


#Start a figure object
fig = plt.figure()

# Define axis (left, bottom, width, height)
axes = fig.add_axes([0,0,1,1])

x = np.linspace(0,5,11)
y = x ** 2

# Color, Linewidth (lw), Alpha (transperancy), Linestyle (ls)
axes.plot(x,y, color='#FF8C00', lw=5, alpha=0.5, ls='-.')

```

**X and Y limits**

```python
#Start a figure object
fig = plt.figure()

# Define axis (left, bottom, width, height)
axes = fig.add_axes([0,0,1,1])

x = np.linspace(0,5,11)
y = x ** 2

# Marker, Marker Size, MarkerFaceColor, MarkerEdgeColor, MarkerEdgeWidth
axes.plot(x,y, ls='', marker='o', markersize=20,
          markerfacecolor='#FF8C00', markeredgecolor='black', markeredgewidth=2
         )

# Set X lower and upper limist
axes.set_xlim([0,3])
# Set Y lower and upper limist
axes.set_ylim([0,10])
```
