## Pandas Builtin Data Vizs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

**Histogram**

```python
h = df1['A'].plot.hist(edgecolor='black', linewidth='1')
h.set_xlim(-5,10)
h.set_ylim(0,350)
```

```python
df1['A'].plot(kind='hist', bins=30)
```

**Area plot**

```python
df2.plot.area(alpha=0.4)
```

**Stacked bar**

```python
df2.plot.bar(stacked=True, alpha=0.6, figsize=[16,6], edgecolor='black', linewidth='1')
```

**Line plot**

```python
df1.plot.line(y='B', figsize=(12,3))
```

**Scatter Plot**

```python
# color based on c column
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm', s=180, figsize=[16,6], edgecolor='black')
```

```python
# s for size, based on column c
df1.plot.scatter(x='A', y='B', s=df1['C']*70, c='C', figsize=[16,6])
```

**Boxplot**

```python
df2.plot.box()
```

**Hex Plot**

```python
df = pd.DataFrame(np.random.rand(1000,2), columns=['a','b'])
```

```python
df.plot.hexbin(x='a', y='b', gridsize=17, figsize=[8,6], cmap='coolwarm')
```

**KDE**

```python
# df2['a'].plot.density()
df2['a'].plot.kde()
```
