# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# To see plots inline in Jupyter
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

**Creating a Plot**  
Funcitonal method


```python
# Outside jupyter you woud need plt.show() to see the graph

# Set x, y, line color and stile
plt.plot(x,y,'r-')
# x label
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

**Mulitple plots in the same canvas**  
Functional method

```python
# 1 row, 2 columns, currently refering to plot 1
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.plot(y,x,'b')

# Refering to plot 2
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

**Creating a plot**  
Object Oriented method

```python
#Start a figure object
fig = plt.figure()

# Define axis (left, bottom, width, height)
axes = fig.add_axes([0.1,0.1,0.8,0.8])

# Plot
axes.plot(x,y)

# Set Labels, Title on the Axes object
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
#Start a figure object
fig = plt.figure()

# Define axis (left, bottom, width, height)
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([1,0.1,0.8,0.8])

# Plot
axes.plot(x,y)
axes2.plot(y,x)

# Titles
axes.set_title('Left Plot')
axes2.set_title('Right Plot')
```

**Multiple plots**  
Object Oriented method

```python
# 3 rows, 3 columns
# the axes object becomes a list of matplot axes
fig,axes = plt.subplots(nrows=3, ncols=3)

# prettify layout
plt.tight_layout()
```

```python
# 2 rows, 1 columns
fig,axes = plt.subplots(nrows=2, ncols=1)

# Index 0 plot
axes[0].plot(x,y)
axes[0].set_title('First plot')

# Index 1 plot
axes[1].plot(y,x)
axes[1].set_title('Second plot')

plt.tight_layout()
```

**Set figure size**  
One plot

```python
# figsize 8inch x 2inch
fig = plt.figure(figsize=(8,2))

# Start at (0,0) corner, occupy full width and height
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

**Set figure size**  
Multile plots

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)
plt.tight_layout()
```

**Save Figure**

```python
fig.savefig('my_picture.png')
```

**Legends**

```python
# figsize 8inch x 2inch
fig = plt.figure()

# Start at (0,0) corner, occupy full width and height
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='One')
ax.plot(y,x, label='Two')
#ax.legend(loc='upper left')
#ax.legend(loc='best')
# Or specify coordinates for the bottom left corner of the legend
ax.legend(loc=(0.8,0.5))

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```
